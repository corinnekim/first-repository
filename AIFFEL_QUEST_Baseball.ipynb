```python
# 정답 숫자 5자리 생성
import random

# 0부터 9까지의 숫자를 문자열 리스트로 준비
population = [str(i) for i in range(10)] # ['0', '1', '2', ..., '9']

# 첫 번째 자리를 선택 (0을 제외한 1~9)
first_digit = random.choice('123456789')

# 전체 숫자(population)에서 첫 번째 자리에서 쓴 숫자를 제거
population.remove(first_digit)

# 나머지 9개의 숫자 중에서 4개를 중복 없이 뽑음
# (random.sample은 순서를 섞어서 리스트로 반환합니다)
other_digits = random.sample(population, 4)

# 첫 번째 자리와 나머지 4자리 숫자를 합쳐서 5자리 문자열을 만듦
result_str = first_digit + "".join(other_digits)

# print(f"생성된 5자리 고유 난수: {result_str}")
# print(f"모든 자리가 고유한가요? {len(set(result_str)) == 5}")


while True:
  # 사용자 입력 받기

  input_num = input("5자리 숫자를 입력하세요 : ")

  # 사용자 입력에 오류가 있다면 재입력 요청
  if not input_num.isdigit():
    print("숫자가 아닙니다. 다시 입력하세요.")
    continue
  elif len(input_num) != len(result_str):
    print("5자리가 아닙니다. 다시 입력하세요. ")
    continue
  elif len(set(input_num)) != len(input_num):
    print("중복된 숫자가 존재합니다. 다시 입력하세요.")
    continue

  # strike, ball 변수 생성
  strike = 0
  ball = 0

  # 다섯 자리 숫자 검사
  for i in range(len(input_num)):
    if input_num[i] == result_str[i]: # strike인지 먼저 확인
      strike += 1
    elif input_num[i] in result_str: #strike가 아니면, ball인지 확인
      ball += 1

  # strike, ball 개수 출력
  print("{} strike, {} ball".format(strike, ball))

  # 정답을 맞혔을 경우 프로그램 종료
  if strike == 5:
    print("정답입니다! 프로그램을 종료합니다.")
    break
```
